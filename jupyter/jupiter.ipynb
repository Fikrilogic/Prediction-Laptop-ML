```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize=(10,10))
```

```python
data = np.array([0.9, 0.7, 0.7, 0.8])
x_name = np.array(['K-Nearest Neighbors', 'Decision Tree', 'Gradient Boosting Decision Tree', 'Naive Bayes'])
df = pd.DataFrame({'algorithm': x_name, "accuracy": [int(n*100) for n in data]})
df
sns.set_theme(style='whitegrid')
ax = sns.barplot(data=df, x=x_name, y='accuracy')
for p in ax.patches:
    percentage = '{}%'.format(10 * p.get_height()/10)
    x = p.get_x() + 0.4
    y = p.get_height() + 1
    ax.annotate(percentage, (x, y),ha='center')
plt.xticks(rotation=15)
plt.yticks(range(10,100, 10))
plt.ylabel('percentage')

plt.show()
```

```python
kfold_data= [
    0.7,
    0.8,
    0.6,
    0.9,
]
kfold_turn = [1,2,3,4]

testing = pd.DataFrame({'turn': kfold_turn, 'accuracy': kfold_data})
testing
```

```python
x_label = [1,2,3,4]

ax = sns.lineplot(data=testing, x=testing['turn'], y='accuracy', hue='accuracy', style='accuracy', dashes=True)
for p in ax.patches:
    percentage = '{}%'.format(p.get_y())
    x = p.get_x() + 0.4
    y = p.get_height() + 1
    ax.annotate(percentage, (x, y),ha='center')
plt.xticks(np.linspace(1, 4, 1))
plt.yticks(np.linspace(0, 1, 11))
plt.show()
```
